```python
#| eval: false
import os
import shutil
```

## Introduction

The original format in which AIRCloud was built and stored differs from the one expected by the dataloader. Therefore, this notebook is required to adjust the folder structure.

## Formatting

In the cell below, define: 
* `SOURCE_NAME`: The name of the folder containing AIRCloud
* `DESTINY_NAME`: The name of the folder that will receive the formatted AIRCloud

```python
#| eval: false
SOURCE_NAME = "AIRCloud_V3"
DESTINY_NAME = "AIRCloud_V3_FORMATTED"
```

The following cell creates the folder that will contain the dataset, already organized in the proper structure.

```python
#| eval: false
DESTINY_PATH = os.path.join("/workspace/data/", DESTINY_NAME)
VELODYNE_PATH = os.path.join(DESTINY_PATH, "data_odometry_velodyne/dataset/sequences/")
LABEL_PATH = os.path.join(DESTINY_PATH, "data_odometry_labels/dataset/sequences/")

os.makedirs(DESTINY_PATH, exist_ok=True)
os.makedirs(VELODYNE_PATH, exist_ok=True)
os.makedirs(LABEL_PATH, exist_ok=True)
```

Next, the base dataset is copied to its new destination, now in the proper format for use with the dataloaders.

::: {.callout-warning}
In this process, the dataset will be duplicated. Feel free to delete the base dataset folder if necessary.
:::

```python
#| eval: false
SOURCE_DATA = os.path.join("/workspace/data/", SOURCE_NAME, "dataset/sequences/")
SEQUENCES = ["07", "08", "09", "10"]

for sequence in SEQUENCES:
    shutil.copytree(
        os.path.join(SOURCE_DATA, sequence, "velodyne"),
        os.path.join(VELODYNE_PATH, sequence, "velodyne")
    )
    shutil.copytree(
        os.path.join(SOURCE_DATA, sequence, "labels"),
        os.path.join(LABEL_PATH, sequence, "labels")
    )
```

Finally, add the `semantic-kitti.yaml` of the corresponding dataset into the formatted dataset folder, alongside `data_odometry_labels` and `data_odometry_velodyne`, and the dataset will be ready for use.
